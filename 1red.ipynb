```python
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))
```

```python
def sigmoid_derivada(x):
    return sigmoid(x)*(1.0-sigmoid(x))
```

```python
#predicion funcion
#prediccion = lambda x, b: step(x + b)
def prediccion(x, b):
    return  sigmoid(x+b)
```

```python
prediccion(1.45, 1)
```

```python
def accuracy(b):
    # corecto = 0
    error_sum = 0
    for i, row in ds_dinos.iterrows():
        guess = prediccion(row.altura, b)
        error = abs(guess - row.agresivo)
        error_sum += error
    return error_sum / len(ds_dinos.agresivo)
```
